```python
import numpy as np
import matplotlib.pyplot as plt

M=float(input("Enter a value for M:"))
x=np.linspace(-10,10,100)
y1=M*(x**2)
y2=M*np.sin(x)
plt.figure(figsize=(10,5))
plt.plot(x,y1,color='Yellow',linestyle='-',label=f"y={M}*(x^2)")
plt.plot(x,y2,color='blue',linestyle='--',label=f"y={M}*sin(x)")
plt.ylim(-5,50)
plt.legend()
plt.grid(True)
plt.title(f"Plots of y={M}*(x^2)  and y={M}*sin(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

df=pd.DataFrame(
    {
    'Subject' : ['Cognitive','AI','DAA','EDP','Aptitude'],
    'Score' : [88,69,97,85,89]
    }
)

plt.figure(figsize=(8,5))

colors=sns.color_palette('husl',len(df))
plot=sns.barplot(x='Subject',y='Score',data=df,palette=colors)

for index,row in enumerate(df.iterrows()):
  plt.text(index,row[1].Score+1,str(row[1].Score),ha='center')

plt.title('Subjects Scores')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
```

```python
np.random.seed(102317239)
data=np.random.randn(50)

plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.plot(np.cumsum(data),color='grey')
plt.title('Cumulative sum')
plt.xlabel('Sum')
plt.ylabel('Cumulative value')
plt.grid(True)

plt.subplot(2,2,2)
x=np.arange(50)
y=data+np.random.normal(scale=0.5,size=50)
plt.scatter(x,y,color='green')
plt.title('Scatter plot with noise')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
dataset_url = 'https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv'
df=pd.read_csv(dataset_url)

plt.figure(figsize=(8,5))
sns.lineplot(x=df['month_number'],y=df['total_profit'],marker='o',color='brown')
plt.title('Total profit of all months')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.grid(True)
plt.show()


plt.figure(figsize=(8,5))
for column in df.columns[1:-1]:
  sns.lineplot(x=df['month_number'],y=df[column],label=column)

plt.title('Product Sales over Month')
plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.show()


plt.figure(figsize=(8,5))
df.sum().drop('month_number').plot(kind='bar',color='red')
plt.title('Total Sales and profit by feature')
plt.xlabel('Features')
plt.ylabel('Total Values')
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
```
