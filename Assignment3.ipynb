```python
def categorize_performance(rating):
  if rating >= 4.5:
    return 'Excellent'
  elif rating >= 4.0:
    return 'Good'
  else:
    return 'Average'

employee_data['Performance Category'] = employee_data['Rating'].apply(categorize_performance)
employee_data
```
